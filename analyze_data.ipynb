# Creating the Directory Structure

```python
max_lattice_size = 6
for n in range(2,max_lattice_size+1):
    for m in range(2,max_lattice_size+1):
        for x in  range(1, max_lattice_size*max_lattice_size):
            if(m>=n): #preventing creation of both 2x3 and 3x2 system sizes
                dir = 'data/'+str(n)+'x'+str(m)+'/'+str(x)+'_occupants'
                if not os.path.exists(dir):
                    os.makedirs(dir)
```

# Functions

```python
def get_taus(filepath):
    taus = []
    seeds = []
    myfile = open(filepath, "rt")
    for line in myfile:
        line = line.rstrip("\n")
        if line.startswith('seed'):
            seeds.append(int(line.split(' ')[-1]))
        if line.startswith('tau'):
            taus.append(float(line.split(' ')[-1]))
    return seeds, taus
```
